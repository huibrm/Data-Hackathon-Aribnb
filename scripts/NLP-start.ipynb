```python
data = pd.merge(best_listing, reviews, on = 'listing_id')
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop = set(stopwords.words('english'))
```

```python
good_comments = data.comments

data['word_count'] = data['comments'].apply(lambda x: len(str(x).split(" ")))
```

```python
data['stopwords'] = data['comments'].apply(lambda x: len([x for x in str(x).split() if x in stop])) 
```

```python
data['comments'] = data['comments'].apply(lambda x: " ".join(x.lower() for x in str(x).split()))
```

```python
data['comments'] = data['comments'].str.replace('[^\w\s]','')
```
